# Circuit Playground Express

Presentation to the San Diego Python User Group, November 2020,
by Erik Colban


(Available at: https://github.com/ecolban/SDPUG/tree/master/circuit_playground_express)

This is a short presentation on the Circuit Playground Express (CPX) that I won in a raffle at the October 2020 meeting. Thanks you, San Diego Python User Group and JFrog, who sponsored the prize.

## Source: [Adafruit](https://learn.adafruit.com/adafruit-circuit-playground-express)



Lots of mounted sensors and outputs lets you play without having to buy additional electronic parts and mount them on a breadboard. 

The CPX can be programmed with:
1. MakeCode (similar to Scratch)/Javascript
2. Arduino IDE (or direct ARM coding)
3. **CircuitPython**

## [CircuitPython](https://learn.adafruit.com/welcome-to-circuitpython)
+ Adafruit adaptation of microPython
+ Based on Python 3 optimized for microcontrollers
+ A subset of Python 3
+ Lots of project/code examples for different boards (including the CPX) on Adafruit's web site.
